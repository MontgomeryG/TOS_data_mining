```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming open_market_df is your DataFrame and indc_list is your list of columns

def convert_percentage(perc_str):
    try:
        # Remove '%' and convert to float
        return float(perc_str.replace('%', ''))
    except ValueError:
        # Handle the case where conversion is not possible
        return None

for column in indc_list:
    try:
        # Convert percentage strings to floats for the current column
        open_market_df[column] = open_market_df[column].apply(convert_percentage)

        # Drop rows where conversion resulted in None (if any)
        temp_df = open_market_df.dropna(subset=[column])

        # Plotting
        temp_df[column].plot(
            title=f'{column} during market hours',
            figsize=[20, 10]
        )

        # Show the plot
        plt.show()

    except Exception as e:
        print(f"An error occurred while processing {column}: {e}")
```
