```python
# Bring in some sample output data.
# read in excel file
df = pd.read_excel("../TOS-data-mining/resources/TOS data 2024 01 16 00 10 02.xlsx")

# Save as .csv
df.to_csv('../TOS-data-mining/resources/TOS-test-data.csv', index=True)

df
```

```python
# Clean up df
df["Date"] = pd.to_datetime(df['Date'])
df.dropna(inplace=True)
df.set_index(df["Date"], inplace=True)
df = df.drop(columns=["Date"])
df
```

```python
# Plot some key columns with hvplot

# # Plot the closing price of TSLA
# df_tsla["close"].plot(
#     title="Historial TSLA Closing Prices", 
#     figsize=[20, 10]
# )

column = 'AAPL Last'

AAPL_last_plot = df[column].plot(
    title = column,
    figsize=[20, 10]
)
```

Let's access other important variables given each of the columns. 

```python
summary_stats = df.describe()
summary_stats
```

```python
summary_stats.index
```

```python
stock_dev = summary_stats.loc["std"]
stock_dev
```

```python
stock_dev.plot(kind='bar')
```

```python
# Create a bar plot
stock_dev.plot(kind='bar')

# Set the y-axis to logarithmic scale
plt.yscale('log')

# Adding labels and title for clarity
plt.xlabel('Stock + Indicator')
plt.ylabel('Standard Deviation (log scale)')
plt.title('Logarithmically Scaled Bar Plot of Standard Deviations')  # Replace with your title

plt.show()
```
